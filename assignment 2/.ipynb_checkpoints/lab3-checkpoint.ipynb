```python
from __future__ import division
% matplotlib inline

import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from scipy.integrate import quad


size=(10,8)
```

```python

#functions

def exponential(x):
    return np.exp(x)

def coscos(x):
    return np.cos(np.cos(x))

#a0_exp= (1/(2*np.pi))*(np.e^(2*np.pi)-1)
#def a_nexp(n):

def expcos(x,k):
    return (exponential(x))*np.cos(k*x)

def expsin(x,k):
    return (exponential(x))*np.sin(k*x)

def cccos(x,k):
    return (coscos(x))*np.cos(k*x)

def ccsin(x,k):
    return (coscos(x))*np.sin(k*x)
```

```python
valuerange= np.arange(-2*np.pi, 4*np.pi, 0.1)

plt.figure(1, figsize= size)
plt.grid(True)
plt.xlabel("$x$")
plt.ylabel("$e^x$")
plt.semilogy(valuerange, exponential(valuerange),"ro")
plt.show()
```

```python

plt.figure(2, figsize= size)
plt.grid(True)
plt.plot(valuerange, coscos(valuerange),"bo")
plt.xlabel("$x$")
plt.ylabel("$cos(cos(x))$")
plt.show()
```

```python
# to find the vector of 25 coefficients

#find the coefficients for exponential

coffexp=[]
a0= quad(exponential, 0, 2*np.pi)[0]

a0= a0/(2*np.pi)
coffexp.append(a0)
for i in range(1,26):
    ai= quad(expcos, 0, 2*np.pi, args=(i))[0]
    bi= quad(expsin, 0, 2*np.pi, args=(i))[0]
    coffexp.append(ai)
    coffexp.append(bi)

coffexp= np.array(coffexp)/(np.pi)


plt.figure(3, figsize= size)

plt.subplot(211)
plt.xlabel("x")
plt.ylabel("Vector of coefficients")
plt.title("Fourier series coefficients on semilog scale")
plt.semilogy( coffexp, "ro")

plt.subplot(212)
plt.loglog( coffexp, "ro")
plt.xlabel("x")
plt.ylabel("Vector of coefficients")
plt.title("Fourier series coefficients on loglog scale")

plt.tight_layout()
plt.show()
plt.close()
```

```python
# to find the vector of 25 coefficients

#find the coefficients for exponential

coffcc=[]
a0= quad(coscos, 0, 2*np.pi)[0]

a0= a0/(2*np.pi)
coffcc.append(a0)
for i in range(1,26):
    ai= quad(cccos, 0, 2*np.pi, args=(i))[0]
    bi= quad(ccsin, 0, 2*np.pi, args=(i))[0]
    coffcc.append(ai)
    coffcc.append(bi)

coffcc= abs(np.array(coffcc))/(np.pi)


plt.figure(4, figsize= size)

plt.subplot(211)
plt.xlabel("x")
plt.ylabel("Vector of coefficients")
plt.title("Fourier series coefficients on semilog scale")
plt.semilogy( coffcc, "ro")

plt.subplot(212)
plt.loglog( coffcc, "ro")
plt.xlabel("x")
plt.ylabel("Vector of coefficients")
plt.title("Fourier series coefficients on loglog scale")

plt.tight_layout()
plt.show()
plt.close()
```
