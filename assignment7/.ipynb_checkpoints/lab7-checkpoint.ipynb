```python
from __future__ import division
% matplotlib inline
import sys
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sp
size=(10,8)
```

```python
# defining useful functions
def u(t):
    return 1*(t>=0)
```

```python
# Problem 1
den=[1,1,4.75,2.25,5.625]
num=[1,0.5]
X= sp.lti(num,den)
T= np.linspace(0,50,101)
t,y=sp.impulse(X,None,T)
plt.plot(t,y)
```

```python
w1,S1,phi1=X.bode()
plt.semilogx(w1,S1)
```

```python
num= [1,0.05]
den=[1,0.1,4.502,0.225,5.068]
X= sp.lti(num,den)
T= np.linspace(0,50,101)
t,y=sp.impulse(X,None,T)
plt.plot(t,y)
```

```python
w2,S2,phi2=X.bode()
plt.semilogx(w2,S2,"b-")
plt.semilogx(w1,S1,"g-")
# note the higher quality factor
```

```python
H=sp.lti([1],[1,0,2.25])
o=[]
t=np.linspace(0,100,201)

for i in np.arange(1.4,1.65,0.05):
    ip= (np.cos(i*t))*(np.exp(-0.05*t))*u(t)
    T,op,svec= sp.lsim(H,ip,t)
    plt.figure(figsize=size)   
    plt.plot(T,op)
    plt.show()
    plt.close()
```

```python
den=[10**(-12),10**(-4),1]
num=[1]
X= sp.lti(num,den)
w,S,phi=X.bode()
plt.subplot(211)
plt.semilogx(w,S,"b-")
plt.subplot(212)
plt.semilogx(w,phi,"g-")
# note the higher quality factor
```

```python
t=np.arange(0,10**(-1), 10**(-7))
f= (np.cos(1000*t)-np.cos((10**(6))*t))*u(t)
T,op,svec= sp.lsim(X,f,t)
```
