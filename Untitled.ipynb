```python
from matplotlib import cm
```

```python
N=50
r=0.35
Niter=1500
length=-np.linspace(-0.5,0.5)
X,Y= np.meshgrid(length,length)
phi=np.zeros((N,N))
ii=np.where(X*X+Y*Y<0.35**2)
#print (ii)
phi[ii]=1
plt.figure()
plt.contourf(X,Y,phi,cmap=cm.jet)
plt.show()
plt.figure()
plt.scatter(X,Y,phi)
```

```python
for _  in range(0,Niter):
    phi[1:-1,1:-1]= phi[1:-1,0:-2] + phi[1:-1,2:] + phi[0:-2,1:-1] + phi[2:,1:-1]
    phi=phi*0.25
    phi[:,0]=phi[:,1]
    phi[:,-1]=phi[:,-2]
    phi[0,:]=phi[1,:]
    phi[-1,:]=0
    phi[ii]=1
    
plt.contour(X,Y,phi)
plt.scatter(X,Y,phi,linewidth=2)
figure4=plt.figure()
ax= p3.Axes3D(figure4)
plt.title(r"Trying stuff $\nabla^{2}$")
ax.plot_surface(X,Y,phi,cmap=cm.jet)
```

```python
jy=phi[0:-2,:]-phi[2:,:]
jx=phi[:,0:-2]-phi[:,2:]
x=X[1:-1,1:-1]
y=Y[1:-1,1:-1]
plt.figure(figsize=(10,8))
plt.quiver(x,y,-jy,-jx)
```

```python
length=np.linspace(0,10,101)
X,Y= np.meshgrid(length,length)
ii=np.where((X-5)*(X-5) + (Y-5)*(Y-5) < 4)
#kk= np.where(0.16<(X-9)*(X-9)+(Y-9)*(Y-9)<0.64 )
kk= np.where(np.logical_and(((X-9)*(X-9)+(Y-9)*(Y-9)>0.16 ),(X-9)*(X-9)+(Y-9)*(Y-9)<0.64 ))
phi=np.zeros((101,101))
phi[ii]=2
phi[kk]=1
fig=plt.figure()
ax=p3.Axes3D(fig)
ax.plot_surface(X,Y,phi,cmap=cm.jet)
plt.show()
```
